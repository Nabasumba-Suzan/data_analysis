```python
import pandas as pd

# Load data from Excel
df = pd.read_excel("Mother_Majeri_Primary_School_2025.xlsx", sheet_name="Student Register", header=3)

# Clean and convert into a list of dictionaries
students = []
for _, row in df.iterrows():
    if pd.notna(row["Student Name"]):
        students.append({
            "name": row["Student Name"],
            "expected": int(row["Total Expected\n(UGX)"]),
            "received": int(row["Total Received\n(UGX)"])
        })
```

```python
class_summary = {}

# Step 1: Accumulate expected and received amounts per class
for student in students:
    student_class = student["class"]
    
    # Part 1B Pattern: Initialize nested dictionary if key missing
    if student_class not in class_summary:
        class_summary[student_class] = {"expected": 0, "received": 0}
        
    class_summary[student_class]["expected"] += student["expected"]
    class_summary[student_class]["received"] += student["received"]

# Step 2: Calculate collection rates and format output
output_parts = []
for student_class, totals in class_summary.items():
    rate = (totals["received"] / totals["expected"]) * 100
    output_parts.append(f"{student_class}: Rate={rate:.1f}%")

print(" | ".join(output_parts))
```

```python
student_balances = []

# Step 1: Build list of (name, balance) tuples
for student in students:
    balance = student["expected"] - student["received"]
    student_balances.append((student["name"], balance))

# Step 2: Part 1C Pattern - Sort using a lambda function key
# x[1] references the balance (second item in tuple), reverse=True sorts highest first
student_balances.sort(key=lambda x: x[1], reverse=True)

# Step 3: Get top 3 highest balances
top_3 = student_balances[:3]

# Format output
formatted_top_3 = [f"{name} - {balance}" for name, balance in top_3]
print(", ".join(formatted_top_3))
```

```python
import pandas as pd
from pathlib import Path

# 1. Define dataset
students = [
    # P1A
    {"name": "Faith Nakato", "class": "P1A", "area": "Ntinda", "t1_expected": 600000, "t1_received": 220000},
    {"name": "Daniel Kato", "class": "P1A", "area": "Kireka", "t1_expected": 450000, "t1_received": 450000},
    {"name": "Moses Okello", "class": "P1A", "area": "Ntinda", "t1_expected": 300000, "t1_received": 500000},
    
    # P1B
    {"name": "Peter Ssemwogerere", "class": "P1B", "area": "Bweyogerere", "t1_expected": 450000, "t1_received": 225000},
    {"name": "Sarah Namubiru", "class": "P1B", "area": "Kireka", "t1_expected": 310000, "t1_received": 155000},
    
    # P2A
    {"name": "John Ochieng", "class": "P2A", "area": "Kireka", "t1_expected": 500000, "t1_received": 400000},
    {"name": "Grace Akello", "class": "P2A", "area": "Bweyogerere", "t1_expected": 500000, "t1_received": 400000},
    
    # P2B
    {"name": "David Mukasa", "class": "P2B", "area": "Bweyogerere", "t1_expected": 580000, "t1_received": 312500},
    {"name": "Mary Joan", "class": "P2B", "area": "Ntinda", "t1_expected": 420000, "t1_received": 312500},
    
    # P3A
    {"name": "Alex Kintu", "class": "P3A", "area": "Ntinda", "t1_expected": 550000, "t1_received": 0}
]

df = pd.DataFrame(students)
df["t1_balance"] = df["t1_expected"] - df["t1_received"]

# 2. Risk scoring function (Part 1C logic)
def risk_flag(expected, received):
    balance = expected - received
    if balance > 300000:
        return "High Risk"
    elif balance > 200000:
        return "Medium Risk"
    elif balance > 0:
        return "Low Risk"
    else:
        return "No Risk"

# 3. Apply risk function across row values using axis=1
df["risk_flag"] = df.apply(
    lambda r: risk_flag(r["t1_expected"], r["t1_received"]), axis=1
)

# 4. Create directory and export CSV
Path("data").mkdir(exist_ok=True)
df.to_csv("data/session_practice_output.csv", index=False)
print("Saved to data/session_practice_output.csv")
```
